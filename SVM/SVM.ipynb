```python
from numpy import loadtxt
```

```python
X = loadtxt('/content/drive/MyDrive/Học máy/(SVM + KNN) using TFIDF & Word2vec/dataIntTrain.csv', delimiter=',')
Y = loadtxt('/content/drive/MyDrive/Học máy/(SVM + KNN) using TFIDF & Word2vec/dataIntTest.csv', delimiter=',')
```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
clf = make_pipeline(StandardScaler(),LinearSVC(random_state=0, tol=1e-5))
clf.fit(X, Y)
```

```python
Xtest = loadtxt('/content/drive/MyDrive/Học máy/preprocessing-data/word2vec/dataIntXTest.csv', delimiter=',')
Ytest = loadtxt('/content/drive/MyDrive/Học máy/preprocessing-data/word2vec/dataIntYTest.csv', delimiter=',')
```

```python
from sklearn.metrics import f1_score
pred = clf.predict(Xtest)
```

```python
f1_score(Ytest, pred, average=None,zero_division=True)
```

```python
from sklearn.metrics import recall_score
recall_score(Ytest, pred, average = None,zero_division=True)
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.1)
```

```python
kn = ['linear', 'poly', 'rbf', 'sigmoid']
for k in kn:
  # clf = svm.SVC(kernel=k)
  clf = make_pipeline(StandardScaler(),svm.SVC(kernel=k))
  clf.fit(X_train,y_train)
  y_out = clf.predict(X_val)
  print("Accuracy kernel ",k," :",np.sum(y_out ==y_val)/y_val.shape)
```

```python
kn = ['linear', 'poly', 'rbf', 'sigmoid']
cn = [1,5,10,15,20]
for c in cn:
  print(c)
  # clf = svm.SVC(kernel=k)
  clf = make_pipeline(StandardScaler(),svm.SVC(kernel=kn[2],C=c))
  clf.fit(X_train,y_train)
  y_out = clf.predict(X_val)
  print("Accuracy kernel ",c," :",np.sum(y_out ==y_val)/y_val.shape)
```

```python
clf = make_pipeline(StandardScaler(),svm.SVC(kernel=kn[2],C=10))
clf.fit(X_train,y_train)
y_out = clf.predict(X_val)
print("Accuracy kernel :",np.sum(y_out ==y_val)/y_val.shape)
```

```python
from sklearn.metrics import f1_score
Y_pred = clf.predict(Xtest)
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(Ytest,Y_pred,average=None)*100
```

```python
from sklearn.metrics import precision_score, recall_score

ps = precision_score(Ytest,Y_pred,average=None)*100
```

```python
rc = recall_score(Ytest,Y_pred,average=None)*100
```
